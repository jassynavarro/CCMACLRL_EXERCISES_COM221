## Dataset File

```python
dataset_url = 'https://github.com/robitussin/CCMACLRL_EXERCISES/blob/main/datasets/loan_approval/train.csv?raw=true'
df = pd.read_csv(dataset_url)
```

```python
df.head(30)
```

```python
df.info()
```

```python
df['person_home_ownership'] = df['person_home_ownership'].map({'RENT': 0, 'MORTGAGE': 1,'OWN' : 2, 'OTHER': 3})
df.head(10)
```

```python
df['loan_intent'] = df['loan_intent'].map({'EDUCATION': 0, 'MEDICAL': 1,'PERSONAL' : 2, 'VENTURE': 3, 'DEBTCONSOLIDATION': 4, 'HOMEIMPROVEMENT': 5})
df.head(10)
```

```python
df['loan_grade'] = df['loan_grade'].map({'A': 0, 'B': 1,'C' : 2, 'D': 3, 'E': 4, 'F': 5, 'G': 6})
df.head(10)
```

```python
df['cb_person_default_on_file'] = df['cb_person_default_on_file'].map({'N': 0, 'Y': 1})
df.head(10)
```

```python
df.drop(['id'],axis=1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x = df.drop("loan_status",axis=1)
y = df["loan_status"]

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=1,test_size=0.3)
```
