## LISTS

```python
fruits = ["apple", "mango", "banana", "strawberry", "orange"]
```

```python
for i in fruits:
  print(i)
```

"List Comprehension"

```python
#append
mynewList = [x for x in fruits]
mynewList
```

```python
#with conditions
mynewList = [x for x in fruits if "e" in x]
mynewList
```

```python
#uppercase
mynewList = [x.upper() for x in fruits]
mynewList
```

"Slicing"

```python
#before = start
#after = end
fruits[1:]
```
